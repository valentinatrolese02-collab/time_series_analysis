Do the following:
1.1. Make a time variable, x, such that 2018-Jan has x1 = 2018, 2018-Feb has x2 = 2018 + 1/12,
2018-Mar has x3 = 2018+2/12 etc. and plot the training data versus x.
1.2. Describe the time series in your own words.

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("../data/train.csv")
print(df)

df["time"] = pd.to_datetime(df["time"], utc=True)
df["x"] = df["time"].dt.year + (df["time"].dt.month - 1 )/12
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["x"], df["total"]) # plot total
plt.xlabel("X variable")
plt.ylabel("Total (training data)")
plt.title("Training data vs time variable x")
plt.grid(True)
plt.show()
```
