1.4. Plot the autocorrelation ρ(k) up to nlag = 30 for the coefficient values above.

```python
import numpy as np
import matplotlib.pyplot as plt

phi1 = -0.7
phi2 = -0.2
n_lag = 30

rho = np.zeros(n_lag + 1)
rho[0] = 1
rho[1] = -phi1 / (1 + phi2)

for k in range(2, n_lag + 1):
    rho[k] = -phi1 * rho[k-1] - phi2 * rho[k-2]

plt.stem(range(n_lag + 1), rho)
plt.xlabel("lag k")
plt.ylabel(r"$\rho(k)$")
plt.title("ACF of the AR(2) process")
plt.grid(True)

plt.show()
```
